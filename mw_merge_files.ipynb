```python
import pandas as pd
```

```python
# Job titles loading
job_titles_df = pd.read_csv('./Resources/job_titles.csv')
job_titles_df
```

```python
filenames = job_titles_df['job_titles'].to_list()
filenames
```

```python
filenames =  list(map(lambda x: x.replace(' ', '_'), filenames))
filenames =  list(map(lambda x: "./Resources/individual_job_listings/" + x , filenames))
filenames =  list(map(lambda x: x + "_jobs.csv" , filenames))
filenames
```

```python
column_names = ["Job Title", "Company Name", "Location","Salary","Description", "link"]
jobs_df = pd.DataFrame(columns = column_names)
for filename in filenames:
    jobs_read_df = pd.read_csv(filename)
    jobs_df = pd.concat([jobs_df, jobs_read_df], ignore_index=True)
```

```python
jobs_df
```

```python
jobs_df.to_csv("Resources/mw_job_listings_scraped.csv", index=False)
```
